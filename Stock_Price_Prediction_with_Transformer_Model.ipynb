```python
def read_csv_to_df(filepath, parse_dates=None, index_col=None):
    """
    Reads a CSV file into a pandas DataFrame.
    
    Args:
        filepath (str): Path to the CSV file.
        parse_dates (list or str, optional): Column(s) to parse as datetime.
        index_col (str or int, optional): Column to set as index.
    
    Returns:
        pd.DataFrame: The loaded DataFrame.
    """
    try:
        df = pd.read_csv(filepath, parse_dates=parse_dates, index_col=index_col)
        print(f"Loaded: {filepath} with shape {df.shape}")
        return df
    except FileNotFoundError:
        print(f"File not found: {filepath}")
        return None
    except Exception as e:
        print(f"Error loading {filepath}: {e}")
        return None
```

```python
import pandas as pd # Used for data manipulation and analysis, especially with tabular data.
import numpy as np # Used for numerical operations, especially with arrays and mathematical functions.
import tensorflow as tf # The core TensorFlow library for building and training machine learning models.
from tensorflow import keras # A high-level API for TensorFlow, simplifying model creation and training.
from tensorflow.keras import layers # Provides various neural network layers like Dense, MultiHeadAttention, etc.
from sklearn.preprocessing import MinMaxScaler # Used for scaling features to a specific range (e.g., 0 to 1).
import matplotlib.pyplot as plt # Used for creating static, interactive, and animated visualizations in Python.
import io # Used to work with streams of data, in this case, to read a string as a file.

# --- 1. Load and Prepare the Data ---
df = read_csv_to_df('cache/AAPL_combined_data_2025-07-05.csv', 
                    parse_dates=['date'], # Parse the 'date' column as datetime objects.
                    index_col='date') # Set the 'date' column as the index of the DataFrame.

if df is None:
    raise ValueError("DataFrame is empty. Please check the file path or content.")

# Convert the 'date' column to datetime objects. This is essential for time-series analysis.
df = df.reset_index()
df['date'] = pd.to_datetime(df['date'])
# Set the 'date' column as the DataFrame's index. This makes time-based operations easier.
df.set_index('date', inplace=True)
# Ensure the DataFrame is sorted by date in chronological order. This is critical for time-series data.
df.sort_index(inplace=True)

# Define the target variable (what we want to predict). In this case, it's the 'Close' price.
TARGET = 'Close'
# Define the features (input variables) by selecting all columns except the target.
FEATURES = [col for col in df.columns if col != TARGET]

# --- 2. Normalize Data ---
# Initialize MinMaxScaler for feature scaling. This scales all values to a range between 0 and 1.
# Normalization helps neural networks converge faster and perform better.
scaler = MinMaxScaler()
# Apply the scaler to the feature columns. `fit_transform` calculates the min/max and then scales the data.
df[FEATURES] = scaler.fit_transform(df[FEATURES])

# Create a separate MinMaxScaler specifically for the target variable ('Close' price).
# This is crucial because we will need to inverse-transform the model's predictions
# back to their original dollar values later.
scaler_target = MinMaxScaler()
# Apply the scaler to the target column. Reshape to (-1, 1) is needed because fit_transform expects a 2D array.
df[TARGET] = scaler_target.fit_transform(df[[TARGET]])

# --- 3. Create Sequences for the Transformer ---
# Define the sequence size (also known as look-back period or window size).
# This means we will use the past 60 days of data to predict the next day's 'Close' price.
SEQUENCE_SIZE = 60

# Initialize empty lists to store input sequences (X) and corresponding target values (y).
X, y = [], []
# Iterate through the DataFrame to create sequences.
# The loop runs until there are enough data points left to form a complete sequence and a target.
for i in range(len(df) - SEQUENCE_SIZE):
    # Append a slice of the DataFrame (SEQUENCE_SIZE rows) as an input sequence.
    # `.values` converts the DataFrame slice into a NumPy array.
    X.append(df.iloc[i:i + SEQUENCE_SIZE].values)
    # Append the 'Close' price at the end of the current sequence as the target.
    y.append(df[TARGET].iloc[i + SEQUENCE_SIZE])

# Convert the lists of sequences and targets into NumPy arrays.
# This is required for TensorFlow model input.
X = np.array(X)
y = np.array(y)

# --- 4. Split data into Training, Validation, and Test sets ---
# This is a time-series split, which means data is split chronologically.
# This prevents "lookahead bias" where the model might learn from future data.
# 80% of data for training.
train_size = int(len(X) * 0.8)
# 10% of data for validation.
val_size = int(len(X) * 0.1)

# Split X and y into training, validation, and test sets based on calculated sizes.
X_train, y_train = X[:train_size], y[:train_size]
X_val, y_val = X[train_size:train_size + val_size], y[train_size:train_size + val_size]
X_test, y_test = X[train_size + val_size:], y[train_size + val_size:]

# Print the shapes of the resulting datasets to verify the split.
print("Data shapes:")
print("X_train:", X_train.shape) # (number of training sequences, sequence length, number of features)
print("y_train:", y_train.shape) # (number of training sequences,)
print("X_val:", X_val.shape)     # (number of validation sequences, sequence length, number of features)
print("X_test:", X_test.shape)   # (number of test sequences, sequence length, number of features)

# --- 5. Build the Transformer Model ---

def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0):
    """
    Creates a single Transformer Encoder block.

    Args:
        inputs (tf.Tensor): The input tensor to the encoder block.
        head_size (int): The dimensionality of the attention heads.
        num_heads (int): The number of attention heads.
        ff_dim (int): The hidden dimension of the feed-forward network.
        dropout (float): Dropout rate for attention and feed-forward layers.

    Returns:
        tf.Tensor: The output tensor of the transformer encoder block.
    """
    # Multi-Head Self-Attention mechanism
    # `inputs` are used for query, key, and value in self-attention.
    x = layers.MultiHeadAttention(
        key_dim=head_size, num_heads=num_heads, dropout=dropout
    )(inputs, inputs)
    # Apply dropout to the attention output.
    x = layers.Dropout(dropout)(x)
    # Add a residual connection (inputs + x) and apply Layer Normalization.
    # Epsilon is a small constant to prevent division by zero.
    x = layers.LayerNormalization(epsilon=1e-6)(inputs + x)

    # Feed Forward Network (FFN) part
    # A 1D convolutional layer (effectively a Dense layer applied independently to each time step).
    ff_output = layers.Conv1D(filters=ff_dim, kernel_size=1, activation="relu")(x)
    # Apply dropout to the FFN output.
    ff_output = layers.Dropout(dropout)(ff_output)
    # Another 1D convolutional layer to project back to the original input dimension.
    ff_output = layers.Conv1D(filters=inputs.shape[-1], kernel_size=1)(ff_output)
    # Add another residual connection (x + ff_output) and apply Layer Normalization.
    x = layers.LayerNormalization(epsilon=1e-6)(x + ff_output)
    return x

def build_transformer_model(
    input_shape, head_size, num_heads, ff_dim, num_transformer_blocks, mlp_units, dropout=0, mlp_dropout=0
):
    """
    Builds the complete Transformer model for time series prediction.

    Args:
        input_shape (tuple): The shape of the input sequences (sequence_length, num_features).
        head_size (int): Dimensionality of attention heads.
        num_heads (int): Number of attention heads.
        ff_dim (int): Hidden dimension of the feed-forward network.
        num_transformer_blocks (int): Number of stacked transformer encoder blocks.
        mlp_units (list): List of integers representing the number of units in each Dense layer of the MLP head.
        dropout (float): Dropout rate for transformer encoder blocks.
        mlp_dropout (float): Dropout rate for the MLP head.

    Returns:
        keras.Model: The compiled Keras Transformer model.
    """
    # Define the input layer with the specified input shape.
    inputs = keras.Input(shape=input_shape)
    x = inputs

    # Add positional encoding to the input.
    # Positional encoding injects information about the relative or absolute position
    # of the tokens in the sequence, which is crucial for transformers as they lack recurrence.
    # `tf.range` creates a sequence of integers from 0 to `input_shape[0]` (sequence length - 1).
    positions = tf.range(start=0, limit=input_shape[0], delta=1)
    # An Embedding layer learns a vector representation for each position.
    pos_embedding = layers.Embedding(input_dim=input_shape[0], output_dim=input_shape[1])(positions)
    # Add the positional embedding to the input features.
    x += pos_embedding

    # Stack multiple transformer encoder blocks.
    for _ in range(num_transformer_blocks):
        x = transformer_encoder(x, head_size, num_heads, ff_dim, dropout)

    # Global Average Pooling 1D: Reduces the sequence dimension by averaging across all time steps.
    # This converts the sequence output of the transformer blocks into a fixed-size vector.
    x = layers.GlobalAveragePooling1D(data_format="channels_last")(x)

    # MLP (Multi-Layer Perceptron) head for the final prediction.
    # This part processes the aggregated features from the transformer blocks.
    for dim in mlp_units:
        # A Dense (fully connected) layer with ReLU activation.
        x = layers.Dense(dim, activation="relu")(x)
        # Apply dropout to prevent overfitting.
        x = layers.Dropout(mlp_dropout)(x)

    # Final output layer: A single Dense neuron with no activation function for regression (predicting a continuous value).
    outputs = layers.Dense(1)(x)

    # Create the Keras Model by specifying its inputs and outputs.
    return keras.Model(inputs=inputs, outputs=outputs)

# Model Hyperparameters: These values are chosen based on experimentation or common practices.
HEAD_SIZE = 128 # Size of each attention head.
NUM_HEADS = 4 # Number of attention heads.
FF_DIM = 4 # Dimension of the feed-forward network inside the transformer block.
NUM_TRANSFORMER_BLOCKS = 4 # Number of stacked transformer encoder layers.
MLP_UNITS = [128] # Architecture of the MLP head (one hidden layer with 128 units).
DROPOUT = 0.25 # Dropout rate for the transformer encoder layers.
MLP_DROPOUT = 0.4 # Dropout rate for the MLP head.

# Instantiate the transformer model with the defined hyperparameters and input shape.
model = build_transformer_model(
    input_shape=(SEQUENCE_SIZE, X_train.shape[2]), # Input shape: (sequence_length, number_of_features)
    head_size=HEAD_SIZE,
    num_heads=NUM_HEADS,
    ff_dim=FF_DIM,
    num_transformer_blocks=NUM_TRANSFORMER_BLOCKS,
    mlp_units=MLP_UNITS,
    dropout=DROPOUT,
    mlp_dropout=MLP_DROPOUT,
)

# Compile the model.
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-4), # Adam optimizer with a specific learning rate.
    loss="mean_squared_error" # Mean Squared Error (MSE) is a common loss function for regression tasks.
)

# Print a summary of the model's architecture, including layer types, output shapes, and number of parameters.
model.summary()

# --- 6. Train the Model ---

# Callbacks are functions that can be applied at certain stages of the training procedure.

# ModelCheckpoint callback: Saves the model weights (or the entire model) during training.
# 'save_best_only=True' ensures that only the model with the best validation loss is saved.
# 'monitor='val_loss'' specifies that validation loss should be monitored for improvement.
checkpoint_callback = keras.callbacks.ModelCheckpoint(
    'best_transformer_model.keras', save_best_only=True, monitor='val_loss'
)
# EarlyStopping callback: Stops training when a monitored quantity has stopped improving.
# 'patience=10' means training will stop if validation loss doesn't improve for 10 consecutive epochs.
# 'restore_best_weights=True' ensures that the model weights from the epoch with the best
# monitored quantity (val_loss) are restored at the end of training.
early_stopping_callback = keras.callbacks.EarlyStopping(
    patience=10, restore_best_weights=True
)

# Train the model using the training data and validate it with the validation data.
history = model.fit(
    X_train, y_train, # Training input features and target values.
    validation_data=(X_val, y_val), # Validation input features and target values.
    epochs=100, # Maximum number of training iterations.
    batch_size=32, # Number of samples per gradient update.
    callbacks=[checkpoint_callback, early_stopping_callback] # List of callbacks to use during training.
)

# --- 7. Evaluate and Visualize Results ---

# Load the best model that was saved during training by the ModelCheckpoint callback.
# This ensures we use the model that performed best on the validation set.
best_model = keras.models.load_model('best_transformer_model.keras')

# Make predictions on the unseen test set using the best trained model.
# The predictions will be in the scaled [0, 1] range.
test_predictions_scaled = best_model.predict(X_test)

# Invert the scaling of the predictions to get them back into their original dollar values.
# `scaler_target` was specifically trained on the 'Close' price.
test_predictions = scaler_target.inverse_transform(test_predictions_scaled)
# Also invert the scaling of the actual test target values for comparison.
y_test_actual = scaler_target.inverse_transform(y_test.reshape(-1, 1)) # Reshape to 2D for inverse_transform.

# Plot the actual vs. predicted stock prices.
plt.figure(figsize=(16, 8)) # Set the figure size for better visualization.
plt.plot(y_test_actual, color='blue', label='Actual Stock Price') # Plot actual prices in blue.
plt.plot(test_predictions, color='red', linestyle='--', label='Predicted Stock Price') # Plot predictions in red dashed line.
plt.title('AAPL Stock Price Prediction using Transformer') # Title of the plot.
plt.xlabel('Time (Days in Test Set)') # X-axis label.
plt.ylabel('Stock Price ($)') # Y-axis label.
plt.legend() # Display the legend for labels.
plt.grid(True) # Add a grid to the plot for better readability.
plt.show() # Display the plot.

# Calculate and print evaluation metrics to quantify model performance.
# Mean Squared Error (MSE): Measures the average of the squares of the errors.
# It gives higher weight to larger errors.
mse = np.mean((test_predictions - y_test_actual)**2)
# Mean Absolute Error (MAE): Measures the average of the absolute differences between predictions and actual values.
# It is less sensitive to outliers than MSE.
mae = np.mean(np.abs(test_predictions - y_test_actual))

print(f"Mean Squared Error (MSE) on Test Set: {mse:.4f}")
print(f"Mean Absolute Error (MAE) on Test Set: {mae:.4f}")
```
